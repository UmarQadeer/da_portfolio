```python
inp3 = inp2.copy()
inp3.mnth[inp3.mnth.isin([5,6,7,8,9])] = 5
np.unique(inp3.mnth)
```

```python
inp3.head()
```

```python
inp3['mnth'].unique()
```

```python
inp3.hr[inp3.hr.isin([0,1,2,3,4,5])] = 0
inp3.hr[inp3.hr.isin([11,12,13,14,15])] = 11
```

```python
inp3['hr'].unique()
```

```python
cat_cols = ['season', 'weathersit', 'weekday', 'mnth', 'hr']
inp3 = pd.get_dummies(inp3, columns=cat_cols, drop_first=True)
```

```python
inp3.head()
```

```python
from sklearn.model_selection import train_test_split

# Perform train-test split
df_train, df_test = train_test_split(inp3, test_size=0.3, random_state=100)

# Display the sizes of the train and test sets
print("Train set size:", len(df_train))
print("Test set size:", len(df_test))
```

```python
y_train = df_train.pop("cnt")
X_train = df_train

y_test = df_test.pop("cnt")
X_test = df_test
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
lr.score(X_train, y_train)
```

```python
# Reporting r2 for the model 
from sklearn.metrics import r2_score
y_train_pred= lr.predict(X_train)
a=r2_score(y_train, y_train_pred)
a=a*100
print("R2 Score:"+"{:.2f}%".format(a))
```

```python
y_test_pred= lr.predict(X_test)
b=r2_score(y_test, y_test_pred)
b=b*100
print("R2 Score:"+"{:.2f}%".format(b))
```
