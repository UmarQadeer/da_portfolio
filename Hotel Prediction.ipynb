```python
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib.figure import Figure
from matplotlib import pyplot
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df=pd.read_csv("data[1].csv")
pd.set_option('display.max_columns',None)
```

```python
df.head(60)
```

```python
cc = pd.read_csv(r'Country-Code[1].csv')
```

```python
df=pd.merge(df,cc,on='Country Code',how='left')
df.head()
```

```python
df.columns = df.columns.str.replace(' ','_')
df.columns
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
#Since there were only 9 records without cuisines, we have replace the null values with Others.
df['Cuisines'].fillna('Others',inplace=True)
```

```python
df.isnull().sum()
```

```python
cntry_dist = df.groupby(['Country']).agg(Count = ('Restaurant_ID','count'))
cntry_dist = cntry_dist.sort_values(by='Count',ascending=False)
cntry_dist
#We observe that India has then highest number of restaurants with 8651 restaurants and USA is number 2 with 434 restaurants
```

```python
cntry_dist.plot(kind='pie',y='Count',figsize=[7,7])
plt.tight_layout() 
```

```python
cntry_dist.plot(kind='bar')
```

```python
city_dist = df.groupby(['Country','City']).agg(Count = ('Restaurant_ID','count'))
city_dist.describe()
#city with max restaurant has count = 5473
#city with min restaurant has count = 1
```

```python
city_dist.sort_values(by='Count',ascending=False)
# we see that new Delhi has the maximum restaurant with 5473
# we observe that multiple cities have only one restaurant.
```

```python
min_cnt_rest = city_dist[city_dist['Count']==1]
min_cnt_rest
#There are 46 cities in 7 different countries with 1 restaurants
```

```python
df.head()
```

```python
labels = 'No Order Del','Order Delivery'
G1 = df[['OD_No','OD_Yes']].value_counts()
chart=G1.plot(kind='pie',autopct='%1.1f%%',explode=[0.1,0 ],labels=labels,shadow=True,startangle=90)

plt.title('Distribution of Order Delivery')

# Show the graph
plt.show()
```

```python
# Percent of rest offering online delivery: 25.7%
# Percent of rest not offering online delivery:74.3%
```

```python
# Assuming 'Restaurant Name', 'OD_Yes', and 'OD_No' are columns in your DataFrame 'df'

# Filter the DataFrame for hotels that accept order delivery 
od_yes_df = df[df['OD_Yes'] == 1]
od_yes_count = od_yes_df['Restaurant_Name'].count()

# Filter the DataFrame for hotels that do not accept order delivery 
od_no_df = df[df['OD_No'] == 1]
od_no_count = od_no_df['Restaurant_Name'].count()

# Create a bar chart
plt.bar(['OD_Yes', 'OD_No'], [od_yes_count, od_no_count])

# Add labels to the bars
plt.text(0, od_yes_count, str(od_yes_count), ha='center', va='bottom')
plt.text(1, od_no_count, str(od_no_count), ha='center', va='bottom')

# Add a title and axis labels
plt.title('Hotels Accepting and Not Accepting Orders Delivery')
plt.xlabel('Categories')
plt.ylabel('Number of Hotels')

# Show the graph
plt.show()
```

```python
G2 = df[['TB_Yes','TB_No']].value_counts()
labels ='No Table Booking','Table Booking'
chart=G2.plot(kind='pie',autopct='%1.1f%%',explode=[0, 0.1],labels=labels,shadow=True,startangle=90)
plt.title('Distribution of Table Booking')

# Show the graph
plt.show()
```

```python
# Percent of rest offering table booking: 12.1%
# Percent of rest not offering table booking: 87.9%
```

```python
# Assuming 'Restaurant Name', 'TB_Yes', and 'TB_No' are columns in your DataFrame 'df'

# Filter the DataFrame for hotels that accept table bookings
tb_yes_df = df[df['TB_Yes'] == 1]
tb_yes_count = tb_yes_df['Restaurant_Name'].count()

# Filter the DataFrame for hotels that do not accept table bookings
tb_no_df = df[df['TB_No'] == 1]
tb_no_count = tb_no_df['Restaurant_Name'].count()

# Create a bar chart
plt.bar(['TB_Yes', 'TB_No'], [tb_yes_count, tb_no_count])

# Add labels to the bars
plt.text(0, tb_yes_count, str(tb_yes_count), ha='center', va='bottom')
plt.text(1, tb_no_count, str(tb_no_count), ha='center', va='bottom')

# Add a title and axis labels
plt.title('Hotels Accepting and Not Accepting Table Bookings')
plt.xlabel('Categories')
plt.ylabel('Number of Hotels')

# Show the graph
plt.show()
```
